```r
suppressMessages(library(Seurat))
suppressMessages(library(dplyr))
library(scuttle)
library(scales)
library(RColorBrewer)
library(ggrepel)
library(tidydr)
```

```r
library(ggsci)
library("scales")
pal<-pal_igv("default")(18)
```

```r
obj<-readRDS(file = "./data/modle/human_anno.rds")
```

```r
color=c(
"Fib"="#8b4a4b",
"SMC"='#1f78b4',
"mSMC"="#A1CFFA",
"others"='lightgrey')
```

```r
obj$label<-ifelse(obj$celltype=="Fib","Fib",
                 ifelse(obj$celltype=="SMC","SMC",
                       ifelse(obj$celltype=="mSMC","mSMC","others"
                           
                       )))
```

```r
table(obj$label)
Idents(obj)<-obj$label
```

```r
obj2<-subset(obj,idents='others',invert=T)
```

```r
table(obj2$label)
```

```r
color=c(
"Fib"="#8b4a4b",
"SMC"='#1f78b4',
"mSMC"="#A1CFFA")
genes<-c("PI16","COL15A1","COL1A1","FN1")
width=12
height=3
Idents(obj2)<-obj2$type
obj_AC<-subset(obj2,idents='AC')
options(repr.plot.width = width, repr.plot.height=height)
Idents(obj_AC)<-obj_AC$label
Idents(obj_AC)<-factor(Idents(obj_AC), levels= c("SMC","mSMC","Fib"))
p<-print(VlnPlot(obj_AC,features = genes,cols = color,ncol = 4,pt.size = 0))

ggsave(p, filename="GSE159677_human_anno_fib_markers_vlnplot_AC.pdf.pdf", path=figuredir,width = width,height = height)
```
