```r

for(s in Sys.glob(file.path(drds,"*_uc.RDS"))){
    key <- gsub("_uc.RDS","",basename(s))
    sp <- readRDS(paste0(drds,key,"_uc.RDS"))
    tissue <- unique(sp@meta.data$tissue)
    stage <- unique(sp@meta.data$stage_ann)
    filter <- unique(sp@meta.data$sample)
    stage_clinical <- unique(sp@meta.data$stage_clinical)
    sp <- NormalizeData(sp)
    sp1 <- AddModuleScore(sp,list(gs$plaque_gene))
    colnames(sp1@meta.data) <- str_replace(colnames(sp1@meta.data),"Cluster1","plaque_score")
    p <- SpatialFeaturePlot(sp1,features = "plaque_score",pt.size.factor = 0.65,min.cutoff = 0.1)+ggplot2::scale_fill_continuous(low ="#54BAD3",high = "#ce1256")+coord_cartesian(xlim = c(0, 22000), ylim = c(0,22000))+
    print(p)
}
```
