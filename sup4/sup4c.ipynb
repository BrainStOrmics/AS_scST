```r
####3B
library(ggplot2)
mer <- readRDS("3b.RDS")
order <- c('W20_N','W0','W2','W6','W10','W20')
week <- order
mer$week <- factor(mer$week,levels = order)
cols <- c('W0'= '#FFE3F0','W2'='#EFA5B9','W6'='#E4D7FF','W10'='#ABA0FB','W20'='#736DC2','W20_N'="#A1CFFA")
```

```r
####density
mer1 <- subset(mer,subset = ann2%in%c("mSMCs_1","mSMCs_2"))
mer2 <- subset(mer1,subset = GC_LC=="LC")
mer3 <- subset(mer2,subset = sample=="save")
mer4 <- subset(mer3,subset = ann2=="mSMCs_1")
mer5 <- subset(mer3,subset = ann2=="mSMCs_2")
```

```r
###joy plot
# ---------------------------------------------
p3 <- ggplot(mer4, aes(x = `Digitization Column`, y = Week,fill=Week,color=Week)) + 
   geom_density_ridges(
    jittered_points = TRUE, position = position_points_jitter(width = 0.2, height = 0.1),
    point_shape = '|',
    alpha = 0.7, scale = 0.8,point_size=0.2,point_alpha=1
  )+
  scale_fill_manual(values = cell_type_cols)+scale_x_continuous(limits = c(-20, 120))+scale_color_manual(values = cell_type_cols)+
  theme_minimal()+
  theme(legend.position = 'None')+
  labs(title = 'mSMCs_1')
#---------------------------------------------
p4 <- ggplot(mer5, aes(x = `Digitization Column`, y = Week,fill=Week,color=Week)) + 
   geom_density_ridges(
    jittered_points = TRUE, position = position_points_jitter(width = 0.2, height = 0.1),
    point_shape = '|',
    alpha = 0.7, scale = 0.8,point_size=0.2,point_alpha=1
  )+
  scale_fill_manual(values = cell_type_cols)+scale_x_continuous(limits = c(-20, 120))+ scale_color_manual(values = cell_type_cols)+
  theme_minimal()+
  theme(legend.position = 'None')+
  labs(title = 'mSMCs_2')
p3
p4
```
